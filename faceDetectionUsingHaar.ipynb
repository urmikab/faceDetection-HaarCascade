```
haarcascade = "/content/haarcascade_frontalface_default.xml"
img = plt.imread("image2.jpg") #img converted to numpy array
plt.imshow(img)
```

```
model=cv2.CascadeClassifier(haarcascade)
faces= model.detectMultiScale(img)
faces # col 1 - x, col2- y, col3 - w, col4 - h
```

```
faces.shape
for face in faces:
    x,y,w,h = face
    cv2.rectangle(img,(x,y),(x+w,y+h),(0,0,0),5)
plt.imshow(img)
```
